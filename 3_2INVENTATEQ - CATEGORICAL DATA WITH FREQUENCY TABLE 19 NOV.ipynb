## TWO WAY FREQUENCY TABLE

```python
pd.crosstab(index=train["clarity"],
           columns=train["color"])
```

```python
# to have no. records in a row
contingency_table = pd.crosstab(
          train["clarity"],
           train["color"],
margins = True
)
contingency_table
```

```python
# To have chisquare test
from scipy.stats import chi2_contingency
chi2_contingency(pd.crosstab(train["clarity"], train["color"]))[0:3]
```
